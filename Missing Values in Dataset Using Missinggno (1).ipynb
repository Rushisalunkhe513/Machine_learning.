# writing function to show missing data in columns

```python
def missing_data(df):
    
    #looking for missing values in dataset
    missing_values = df.isnull().sum()
    
    #calculate missing values in %
    missing_value_percentage = 100 * df.isnull().sum()/len(df)
    
    #making table to showcase missing values and percentage
    miss_value_table = pd.concat([missing_values,missing_value_percentage],axis=1)
    
    # now renaming the column with misssing_values and % of missing values.
    miss_value_table = miss_value_table.rename(
    columns={0:"missing_values",1:"% of_missing_values"})
    
    # Sort the table by percentage of missing descending
    miss_val_table_ren_columns = miss_value_table[
        miss_value_table.iloc[:,1] != 0].sort_values(
        '% of_missing_values', ascending=False).round(1)
    
    # print summary of information
    print("your selected dataset have ", df.shape[1],"\n no of columns and no of missing values from that datset are ", miss_val_table_ren_columns.shape[0]," columns taht have missing values.")
    return miss_val_table_ren_columns
```

```python
missing_data(df)
```

```python
# we can deal with missing data in two ways 
# 1) by deleting the column which will cause lot of missing information
# 2) filling null values by constant or by mean,median,mode

from sklearn.impute import SimpleImputer
# making copy of dataset
df_copy = df.copy()
# applying imputer on copied dataset.
constant_imputer = SimpleImputer(strategy="constant")
# filling all missing values by df_copy.iloc[:,:]
df_copy.iloc[:,:]= constant_imputer.fit_transform(df_copy)
df_copy.isnull().sum()

```
